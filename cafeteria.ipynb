```python
class Persona:
    
    lista=[]

    def __init__(self, nombre, telefono):
        self.nombre=nombre
        self.telefono=telefono

    def registrar(self):
        Persona.lista.append(self) 
        print(f"La persona {self.nombre} ha sido registrada con el número {self.telefono}")

    @classmethod
    def personas_registradas(cls):
        print("Personas registradas")
        for Persona in cls.lista:
            print(f"-{Persona.nombre} - {Persona.telefono} -")
```

```python
class Cliente(Persona):
    historial_pedidos = []
    def __init__(self, nombre, telefono):
        super().__init__(nombre, telefono)
        self.pedidos=[]
        
    def realizar_pedido(self, pedido):
        self.pedidos.append(pedido) #agregar pedido
        Cliente.historial_pedidos.append(pedido) #guardarlo en la lista del historial
        print(f"Pedido realizado por {self.nombre}.")
      
    @classmethod 
    def consultar_historial(cls):
        print("Historial de pedidos:")
        for pedido in cls.historial_pedidos:
            print(f"- Cliente: {pedido.cliente.nombre} -")
            print("Productos:")
            for producto in pedido.productos:
                print(f"-{producto.nombre}")
```

```python
class Empleado(Persona):
    def __init__(self, nombre, telefono, rol):
        super().__init__(nombre,telefono)
        self.rol=rol

    def realizar_tarea(self):
    
        tareas = { #diccionario de metodos de acuerdo al rol
            "mesero": self.tomar_pedido,
            "barista": self.preparar_bebida,
            "gerente": self.administrar_cafeteria
        }
        tarea = tareas.get(self.rol)  # obtiene la tarea segn el rol
        return tarea()

    def tomar_pedido(self):
        return f"{self.nombre} (Mesero) esta tomando un pedido"

    def preparar_bebida(self):
        return f"{self.nombre} (Barista) esta preparando una bebida"

    def administrar_cafeteria(self):
        return f"{self.nombre} (Gerente) esta gestionando la cafeteria"
    
```

```python
class Producto_Base:
    def __init__(self,nombre, precio):
        self.nombre=nombre
        self.precio=float(precio) #convierte siempre a float el precio
```

```python
class Bebida(Producto_Base):
    def __init__(self,nombre, precio,tamaño, tipo,opciones_personalizables=None):
        super().__init__(nombre,precio)
        self.tipo=tipo
        self.tamaño=tamaño
        self.opciones_personalizables=opciones_personalizables if opciones_personalizables else [] #Si opciones_personalizables tiene un valor (!= None), se usa ese valor. Si no se proporciona (None), se asigna una lista vacía [].
    
    def descripcion_bebida(self):
        print(f"Esta bebida es: {self.nombre} tamaño {self.tamaño}, {self.tipo}, y cuesta ${self.precio}")
        return f"Y sus opciones personalizables son : {self.opciones_personalizables}"
```

```python
class Postre(Producto_Base):
    def __init__(self,nombre,precio, vegano, singluten):
        super().__init__(nombre, precio)
        self.vegano=vegano
        self.singluten=singluten
    
```

```python
class Inventario:
    def __init__(self):
        self.ingredientes = {} #se crea un diccionario para gestionar el stock

    def agregar_ingrediente(self, ingrediente, cantidad):
        if ingrediente in self.ingredientes: #verifica si el ingrediente ya esta en el inventario
            self.ingredientes[ingrediente] += cantidad #si el ingrediente ya existe, suma ka cantdad nueva
        else:
            self.ingredientes[ingrediente] = cantidad #si no existe, lo agrega con la cantidad indicada

    def verificar_disponibilidad(self, ingredientes_necesarios): #diccionario
        for ingrediente, cantidad in ingredientes_necesarios.items(): #recorre lista de ingredientes ciclo for
            if self.ingredientes.get(ingrediente,0) < cantidad: #busca el ingrediente en el inventario, si no lo encuentra devuelve 0, y luego compara la cantidad
                return "No hay suficiente stock" #no tiene stock
        return "Si hay disponibilidad de los ingredientes necesarios"  #si tiene stock
    
```

```python
class Pedido:
    estados = ['Pendiente', 'En preparacion', 'Entregado']#atributo de estados, es una lista con los posibles estados del pddido

    def __init__(self, cliente):
        self.cliente = cliente
        self.productos = [] #lista
        self.estados = self.estados[0] #estado inicial, pendiente, porque esta en el elemento 0 de la lista

    def agregar_producto(self,producto):
        self.productos.append(producto)
        print("El producto se agregó con éxito")

    def calcular_total(self):
        return sum(producto.precio for producto in self.productos) #suma todos los productos 
    
    def actualizar_estado(self, nuevo_estado):
        if nuevo_estado in self.estados:
            self.estado = nuevo_estado
```

```python
class Promocion:
    def __init__(self,descuento):
        self.descuento = descuento

    def aplicar_promocion(self,cliente):
        if len(cliente.pedidos) >= 5: #verifica si el cliente tiene al menos 5 pedidos,(len sirve para obtener el numero de elemntos de una lista)
            total = sum(pedido.calcular_total() for pedido in cliente.pedidos)
            descuento = total * self.descuento / 100
            total_con_descuento = total - descuento
            print(f"Has recibido un descuento {cliente.nombre}")
            print(f"Total con descuento: ${total_con_descuento}")
        else:
            print(f"Lo siento, esta promoción no es válida")
```
